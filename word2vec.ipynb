```python
import pandas as pd
from gensim.models import FastText
```

```python
df = pd.read_excel('augmented.xlsx')
print(df.info())
```

```python
sentences = []
def get_tokenized_sentences(text):
    text = text.split()
    sentences.append(text)
```

```python
df['Agr_Class'].value_counts()
```

```python
df['text'].map(lambda x: get_tokenized_sentences(x))
```

```python
# train model
model = FastText(sentences,min_count=1,size=100)
# summarize the loaded model
print(model)
# summarize vocabulary
words = sorted(list(model.wv.vocab))
print(words)
# access vector for one word
print(model['modi'])
# save model
#model.save('model.bin')
# load model
#new_model = Word2Vec.load('model.bin')
#print(new_model)
```

```python
corpus = []
def get_uniques(text):
    txt_words = list(set(text.split()))
    corpus.extend(txt_words) 
    #corpus = list(set(corpus))
```

```python
df['text'].map(lambda x: get_uniques(x))
corpus = set(corpus)
len(corpus)
```
